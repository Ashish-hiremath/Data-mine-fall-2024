```python
import pandas as pd
import os

pd.set_option('display.max_columns', None)

# Directory where the CSV files are located
directory = r'C:\Users\ASH\OneDrive\Desktop\data mine\risk factor\ND\data 2015-june 2024'
# List to hold all DataFrames
dfs = []
i=0
# Loop through all CSV files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        filepath = os.path.join(directory, filename)
        df = pd.read_csv(filepath)
        dfs.append(df)
        print(" --> ", i, end="")
        i += 1

print("\nCombining all DataFrames ...")
# Combine all DataFrames into one
df = pd.concat(dfs, ignore_index=True)

print("All CSV files have been uploaded and combined!")
```

Transform numeric columns into float

```python
columns_to_convert = ['MAGNITUDE', 'BEGIN_RANGE', 'END_RANGE', 'BEGIN_LAT', 'BEGIN_LON', 'END_LAT', 'END_LON']

# Convert columns to float and coerce errors to NaN
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Replace NaN values with the mean of each column
for column in columns_to_convert:
    mean_value = df[column].mean()  # Calculate the mean of the column
    df[column].fillna(mean_value, inplace=True)  # Replace NaN values with the mean

# Check the result
print(df[columns_to_convert].head())  # Display the first few rows of the converted columns
```

### Column Descriptions:


1. **EVENT_ID**: A unique identifier for each recorded event (e.g., storm, hail, etc.).

2. **CZ_NAME_STR**: The name of the county or zone where the event occurred (e.g., "BOONE CO.").

3. **BEGIN_LOCATION**: The location (usually a city or town) where the event began (e.g., "LEBANON").

4. **BEGIN_DATE**: The date when the event started (e.g., "05/04/2015").

5. **BEGIN_TIME**: The time (in local time zone) when the event started (e.g., "1715" for 5:15 PM).

6. **EVENT_TYPE**: The type of weather event that occurred (e.g., "Hail", "Tornado", "Flood").

7. **MAGNITUDE**: The magnitude of the event, specific to the event type (e.g., size of hail in inches, wind speed in mph for tornadoes).

8. **TOR_F_SCALE**: The Fujita scale rating for tornadoes (used to indicate the intensity of the tornado), if applicable.

9. **DEATHS_DIRECT**: The number of direct fatalities caused by the event.

10. **INJURIES_DIRECT**: The number of direct injuries caused by the event.

11. **DAMAGE_PROPERTY_NUM**: The estimated property damage caused by the event, usually represented in numerical value (e.g., dollars).

12. **DAMAGE_CROPS_NUM**: The estimated damage to crops caused by the event, usually represented in numerical value (e.g., dollars).

13. **STATE_ABBR**: The two-letter abbreviation for the state where the event occurred (e.g., "IN" for Indiana).

14. **CZ_TIMEZONE**: The time zone in which the event occurred, including the offset from UTC (e.g., "EST-5").

15. **MAGNITUDE_TYPE**: The type of magnitude measurement used (e.g., inches for hail, mph for wind).

16. **EPISODE_ID**: A unique identifier for the entire episode of events (an episode may contain multiple events).

17. **CZ_TYPE**: The type of zone (e.g., "C" for County, "Z" for Zone).

18. **CZ_FIPS**: The Federal Information Processing Standards (FIPS) code, which uniquely identifies counties or regions in the U.S.

19. **WFO**: The National Weather Service Weather Forecast Office identifier responsible for the event record (e.g., "IND" for Indianapolis).

20. **INJURIES_INDIRECT**: The number of indirect injuries related to the event (e.g., injuries sustained during post-event cleanup).

21. **DEATHS_INDIRECT**: The number of indirect fatalities related to the event (e.g., deaths due to cleanup or secondary impacts).

22. **SOURCE**: The source of the event report (e.g., "Law Enforcement", "Trained Spotter", "Emergency Manager").

23. **FLOOD_CAUSE**: If the event involved flooding, the cause of the flood (e.g., "Heavy Rain", "Storm Surge").

24. **TOR_LENGTH**: The length of the tornado's path (in miles), if applicable.

25. **TOR_WIDTH**: The width of the tornado's path (in yards), if applicable.

26. **BEGIN_RANGE**: The range or distance from a specific reference point to where the event began (often used in aviation or maritime reports).

27. **BEGIN_AZIMUTH**: The azimuth (direction in degrees from north) from a specific reference point to where the event began.

28. **END_RANGE**: The range or distance from a specific reference point to where the event ended.

29. **END_AZIMUTH**: The azimuth (direction in degrees from north) from a specific reference point to where the event ended.

30. **END_LOCATION**: The location (usually a city or town) where the event ended.

31. **END_DATE**: The date when the event ended.

32. **END_TIME**: The time (in local time zone) when the event ended.

33. **BEGIN_LAT**: The latitude of the location where the event began.

34. **BEGIN_LON**: The longitude of the location where the event began.

35. **END_LAT**: The latitude of the location where the event ended.

36. **END_LON**: The longitude of the location where the event ended.

37. **EVENT_NARRATIVE**: A detailed description of the event, including any significant impacts or unusual aspects (e.g., "Hail was covering the ground").

38. **EPISODE_NARRATIVE**: A description of the broader episode of events, usually covering multiple events within a single episode (e.g., "Thunderstorms developed south of a frontal boundary...").

39. **ABSOLUTE_ROWNUMBER**: The absolute row number for tracking records in the dataset.


# Part 1: Exploratory Analysis of Weather Event Types


To start, we will examine the distribution of different weather event types in the dataset. This initial analysis will help us understand which events are the most frequent.

#### Steps:
1. Count the occurrences of each event type (`EVENT_TYPE`).
2. Create a bar chart to visualize the frequency of each event type.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for the plots
sns.set(style="whitegrid")

# Load the data
# Make sure to load your DataFrame, for example: df = pd.read_csv('combined_storm_data.csv')

# Count occurrences of each event type
event_type_counts = df['EVENT_TYPE'].value_counts()

# Display the top 10 most frequent event types
print(event_type_counts.head(10))

# Create a bar chart with enhanced styling
plt.figure(figsize=(12, 7))
sns.barplot(x=event_type_counts.head(10).index, y=event_type_counts.head(10), palette='coolwarm')

# Add customizations
plt.title('Top 10 Weather Event Types', fontsize=16)
plt.xlabel('Event Type', fontsize=14)
plt.ylabel('Number of Occurrences', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Add gridlines for clarity
plt.grid(True, which='both', linestyle='--', linewidth=0.7)

# Add tight layout for better fit
plt.tight_layout()


# Display the plot
plt.show()
```

Based on the bar chart, we can observe the following insights:

1. **Thunderstorm Wind** is by far the most frequent weather event in the dataset, with over 4000 occurrences. This suggests that wind-related events are quite common in the dataset, possibly due to their widespread occurrence across different regions and times of the year.

2. **Hail** and **Winter Weather** are the next most frequent event types, with around 1500 and 1000 occurrences, respectively. These events could be region-specific, particularly in areas prone to hailstorms and cold winter conditions.

3. Other significant events include **Flood**, **Flash Flood**, **Winter Storm**, and **Heavy Snow**, which also appear relatively frequently, but at much lower levels compared to Thunderstorm Wind.

4. **Tornadoes**, **Strong Wind**, and **Extreme Cold/Wind Chill** appear less frequently in comparison, which could suggest they are more localized or happen less frequently.

This distribution provides a clear understanding of the most common types of weather events and allows us to focus on particular event types for deeper analysis.

# Part 2: Analyzing Event Magnitude and Impact


In this part, we'll analyze the magnitude of the events (where applicable) and examine their potential impact, such as the number of injuries and deaths (both direct and indirect). This will give us insights into which event types are not only frequent but also severe.

Steps:
Magnitude: Explore the distribution of the magnitude of events (where applicable, such as hail size or wind speed).
Direct Injuries and Deaths: Visualize the number of direct injuries and deaths caused by different event types.
Indirect Injuries and Deaths: Similarly, analyze indirect injuries and deaths for these events.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter out events with missing or zero magnitude for a cleaner analysis
magnitude_events = df[df['MAGNITUDE'] > 0]

# Plot the distribution of event magnitudes (where applicable)
plt.figure(figsize=(10,6))
sns.histplot(magnitude_events['MAGNITUDE'], bins=30, kde=True)
plt.title('Distribution of Event Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.show()

# Plot total direct injuries and deaths by event type
impact_columns = ['INJURIES_DIRECT', 'DEATHS_DIRECT']
impact_data = df.groupby('EVENT_TYPE')[impact_columns].sum().sort_values(by='INJURIES_DIRECT', ascending=False).head(10)

# Plot injuries and deaths
impact_data.plot(kind='bar', figsize=(10,6))
plt.title('Top 10 Event Types by Direct Injuries and Deaths')
plt.xlabel('Event Type')
plt.ylabel('Total Count')
plt.xticks(rotation=45)
plt.show()

# Plot total indirect injuries and deaths by event type
indirect_impact_columns = ['INJURIES_INDIRECT', 'DEATHS_INDIRECT']
indirect_impact_data = df.groupby('EVENT_TYPE')[indirect_impact_columns].sum().sort_values(by='INJURIES_INDIRECT', ascending=False).head(10)

# Plot indirect injuries and deaths
indirect_impact_data.plot(kind='bar', figsize=(10,6))
plt.title('Top 10 Event Types by Indirect Injuries and Deaths')
plt.xlabel('Event Type')
plt.ylabel('Total Count')
plt.xticks(rotation=45)
plt.show()
```

### Interpretation:

#### 1. Distribution of Event Magnitudes:
   - The distribution of magnitudes shows several peaks, with one particularly high spike around a magnitude of 40. This suggests that many weather events in the dataset have a specific magnitude (perhaps wind speed or another common measure), which is clustered around this value.
   - There is also a noticeable concentration of events with very low magnitude, likely reflecting minor weather events.
   - The distribution appears right-skewed, with fewer events having higher magnitudes.

#### 2. Top 10 Event Types by Direct Injuries and Deaths:
   - **Tornadoes** are the most dangerous events in terms of direct injuries, with over 100 recorded injuries. This indicates that tornadoes have a high impact when they occur.
   - **Thunderstorm Wind** and **Winter Weather** also have a significant impact, with a relatively high number of direct injuries and deaths. These events may be more frequent and widespread, contributing to their higher injury counts.
   - **Flash Flood** and **Rip Current** are also notable for causing a significant number of direct deaths, despite fewer injuries. This highlights the deadly nature of certain weather events, even if they are not as frequent.

#### 3. Top 10 Event Types by Indirect Injuries and Deaths:
   - **Heavy Snow** stands out as the event type causing the most indirect injuries, suggesting that post-event conditions (such as icy roads or heavy snow accumulation) are dangerous even after the event has passed.
   - **Winter Weather** also causes a significant number of indirect injuries and deaths, likely due to hazards like slippery conditions, road accidents, or exposure to cold temperatures.
   - **Lightning**, **Dense Fog**, and **Flash Floods** are also associated with some indirect injuries and deaths, underscoring the wide variety of weather conditions that can pose indirect risks to human safety.

This analysis highlights not just the frequency of weather events but their severity in terms of injuries and deaths. Tornadoes, despite being less frequent, are particularly dangerous, while other events like Heavy Snow and Winter Weather pose substantial indirect risks.


# Part 3: Analyzing Property and Crop Damage


In this section, we will focus on the financial impact of weather events by analyzing the damage to property and crops. This can give us insights into which events cause the most economic damage, which is a crucial aspect of understanding the severity of different weather events.

#### Steps:
1. **Property Damage**: Analyze the total property damage caused by different event types.
2. **Crop Damage**: Analyze the total crop damage caused by different event types.
3. **Visualization**: Create bar charts to visualize the top event types that cause the most property and crop damage.

```python
# Convert the damage columns to numeric values (assuming the data contains numerical values)
df['DAMAGE_PROPERTY_NUM'] = pd.to_numeric(df['DAMAGE_PROPERTY_NUM'], errors='coerce')
df['DAMAGE_CROPS_NUM'] = pd.to_numeric(df['DAMAGE_CROPS_NUM'], errors='coerce')

# Group by event type to calculate total damage
damage_columns = ['DAMAGE_PROPERTY_NUM', 'DAMAGE_CROPS_NUM']
damage_data = df.groupby('EVENT_TYPE')[damage_columns].sum().sort_values(by='DAMAGE_PROPERTY_NUM', ascending=False).head(10)

# Plot total property and crop damage
damage_data.plot(kind='bar', figsize=(10,6))
plt.title('Top 10 Event Types by Property and Crop Damage')
plt.xlabel('Event Type')
plt.ylabel('Total Damage (in dollars)')
plt.xticks(rotation=45)
plt.show()
```

#### Property and Crop Damage:
- **Tornadoes** cause the most significant property damage by a large margin, with over 120 million dollars in damage. This emphasizes the destructive power of tornadoes in terms of infrastructure and housing.
- **Flash Floods** and **Thunderstorm Winds** follow, with each causing significant damage to property, but much less in comparison to tornadoes.
- **Floods** also contribute to considerable property damage, although their economic impact is smaller compared to tornadoes and flash floods.
- Crop damage appears to be much smaller overall compared to property damage, with a few events like **Flood** and **Heavy Snow** causing minor crop damage.
- Events like **Lightning**, **High Wind**, and **Winter Weather** are also responsible for some property damage, though their impact is much smaller.

This analysis shows that tornadoes, flash floods, and thunderstorm winds are the most destructive weather events in terms of economic damage, particularly for property.

# Part 4: Correlation Between Event Attributes

In this deeper part of the analysis, we will explore potential relationships between different numerical attributes in the dataset. By calculating and visualizing the correlations, we can identify significant relationships that might reveal patterns or help in modeling.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select relevant numerical columns for correlation analysis
correlation_columns = ['MAGNITUDE', 'DEATHS_DIRECT', 'INJURIES_DIRECT', 'DAMAGE_PROPERTY_NUM',
                       'DAMAGE_CROPS_NUM', 'DEATHS_INDIRECT', 'INJURIES_INDIRECT']

# Compute the correlation matrix
correlation_matrix = df[correlation_columns].corr()

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Weather Event Attributes')
plt.show()
```

### Interpretation:

The correlation matrix heatmap provides several interesting insights into the relationships between the different weather event attributes.

#### Key Findings:

1. **Injuries Directly Correlate with Property Damage**:
   - There is a **strong positive correlation (0.43)** between **Injuries Direct** and **Property Damage**. This suggests that events causing a large number of injuries tend to also result in significant property damage. This could indicate that more severe events not only harm people but also cause widespread destruction to infrastructure.

2. **Moderate Correlation Between Direct Deaths and Property Damage (0.24)**:
   - Direct deaths are moderately correlated with property damage, which is somewhat expected for major events like tornadoes and flash floods. While the correlation isn’t as high as for injuries, it indicates that events causing direct fatalities tend to also involve substantial property damage.

3. **Weak or No Correlation with Magnitude**:
   - Surprisingly, **Magnitude** shows very little correlation with any of the other variables, including **Deaths**, **Injuries**, and **Damages**. This indicates that the magnitude of a weather event (such as wind speed or hail size) does not necessarily translate into higher human impact or economic damage. This could be due to the variety of event types being measured on different scales (e.g., wind speed for hurricanes vs. hail size for hailstorms).

4. **Crop Damage is Isolated**:
   - **Crop Damage** appears to have virtually no correlation with other variables, including **Property Damage**, **Injuries**, or **Deaths**. This suggests that crop damage is likely caused by specific types of weather events (like droughts or floods) that may not necessarily have the same impact on human life or infrastructure.

5. **Indirect Injuries and Deaths Show Little to No Correlation**:
   - Both **Indirect Injuries** and **Indirect Deaths** show very weak correlations with the other variables, which might imply that post-event hazards leading to these injuries or deaths (such as road accidents or long-term effects) are not closely tied to the immediate severity of the event.

#### Insights:
- The **strong correlation** between direct injuries and property damage can help focus attention on events where human safety and economic loss are both major concerns. These are likely to be the events we should prioritize for early warning systems and disaster preparedness.
- The **lack of correlation with magnitude** is an interesting finding. This suggests that focusing on event magnitude alone may not be sufficient to predict its impact. We might need to look at other event characteristics or context, such as population density or infrastructure resilience, to understand the full impact of an event.

This deeper correlation analysis reveals important relationships and highlights areas where further investigation might be needed.


# Part 5: Time Series Analysis of Weather Events


In this section, we will perform a time series analysis to understand how the frequency and impact of weather events have changed over time. This will help us identify trends, such as whether certain types of events have become more frequent or if their impact (in terms of injuries, deaths, or damages) has increased over time.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for the plots
sns.set(style="whitegrid")

# Convert the BEGIN_DATE column to datetime format if not already done
df['BEGIN_DATE'] = pd.to_datetime(df['BEGIN_DATE'], errors='coerce')

# Extract the year from the date for time series analysis
df['YEAR'] = df['BEGIN_DATE'].dt.year

# Group by year to calculate event frequency and total impact metrics
event_trends = df.groupby('YEAR').agg({
    'EVENT_ID': 'count',  # Count number of events
    'DEATHS_DIRECT': 'sum',
    'INJURIES_DIRECT': 'sum',
    'DAMAGE_PROPERTY_NUM': 'sum',
    'DAMAGE_CROPS_NUM': 'sum'
}).reset_index()

# Rename columns for better clarity
event_trends.columns = ['Year', 'Number of Events', 'Direct Deaths', 'Direct Injuries', 'Property Damage', 'Crop Damage']

# Plot the trend of event frequency over the years with customization
plt.figure(figsize=(12, 7))
plt.plot(event_trends['Year'], event_trends['Number of Events'], color='dodgerblue', marker='o', linestyle='-', linewidth=2, markersize=6, label='Number of Events')
plt.title('Trend of Weather Event Frequency Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Events', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()

# Plot the trend of injuries, deaths, and property damage over the years with enhanced styling
plt.figure(figsize=(12, 7))
plt.plot(event_trends['Year'], event_trends['Direct Deaths'], color='firebrick', marker='o', linestyle='-', linewidth=2, markersize=6, label='Direct Deaths')
plt.plot(event_trends['Year'], event_trends['Direct Injuries'], color='darkorange', marker='s', linestyle='--', linewidth=2, markersize=6, label='Direct Injuries')
plt.plot(event_trends['Year'], event_trends['Property Damage'], color='seagreen', marker='^', linestyle='-.', linewidth=2, markersize=6, label='Property Damage (in dollars)')
plt.title('Trend of Event Impact Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Impact', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

### Interpretation:

#### 1. Trend of Weather Event Frequency Over Time:
   - The chart shows a fluctuating trend in the number of weather events between 2015 and 2024.
   - Notably, there was a peak in the number of events around **2019**, followed by a sharp decline in 2020. This could indicate a particularly active year for weather events followed by a less intense year.
   - Another peak can be observed in **2022**, and the frequency of events seems to have dropped sharply again in 2024.
   - These fluctuations suggest that while the number of events can vary significantly from year to year, there might be external factors (e.g., climate conditions, data collection) contributing to these patterns.

#### 2. Trend of Event Impact Over Time:
   - **Property damage** (in green) shows a significant increase in 2021 and again in 2023. These peaks could be tied to specific large-scale weather events that caused massive damage, such as hurricanes, floods, or tornadoes.
   - **Direct injuries** (in orange) and **direct deaths** (in red) remain relatively low and stable over the years. This could suggest that while weather events might be increasing in frequency, the direct human toll has not increased proportionally, possibly due to better disaster preparedness, early warnings, or improved infrastructure.
   - However, it is clear that while the human toll remains relatively flat, **economic losses** (property damage) have become much more significant, especially in recent years. This may reflect the increasing cost of weather-related damages due to urbanization, expensive infrastructure, or more extreme weather events.

#### Insights:
- The trends indicate that **2021** and **2023** were particularly destructive years in terms of property damage. These years likely experienced one or more significant weather events that led to these spikes.
- While human fatalities and injuries remain relatively stable, the **economic impact of weather events has escalated**, highlighting the growing financial risks posed by natural disasters.
- This suggests that future preparedness efforts might need to focus not only on minimizing human casualties but also on reducing the economic impact of these events, possibly through stronger building codes, better infrastructure planning, and increased insurance coverage.
